```python
import zipfile
import os
import cv2
import numpy as np
from io import BytesIO

# Pre-define the image size
def resize_image(image, target_size=(224, 224)):
    return cv2.resize(image, target_size)

# process the image from the zip
def process_images_in_zip(zip_path, output_folder):


    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the zip
    with zipfile.ZipFile(zip_path, 'r') as archive:
        # get all the image
        image_files = [f for f in archive.namelist() if f.endswith(('.jpg', '.jpeg', '.png'))]

        total_images = len(image_files)
        print(f"Total images to process: {total_images}")

        # loop through the picture
        for i, image_file in enumerate(image_files):
            # read the file
            with archive.open(image_file) as file:
                image_data = file.read()


            image_array = np.frombuffer(image_data, np.uint8)
            color_image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)


            if color_image is None:
                print(f"Skipping non-image file: {image_file}")
                continue


            resized_image = resize_image(color_image, target_size=(224, 224))


            output_image_path = os.path.join(output_folder, os.path.basename(image_file))
            cv2.imwrite(output_image_path, resized_image)


            print(f"Processed {i + 1}/{total_images} images ({((i + 1) / total_images) * 100:.2f}%)")
```

```python

# ZIP 文件路径
zip_path = '/content/drive/MyDrive/CapstoneProject/data/train.zip'  # 替换为你的 ZIP 文件路径
output_folder = '/content/drive/MyDrive/CapstoneProject/data/ColorTrain'  # 输出文件夹路径，保存处理后的图像

# 开始处理 ZIP 文件中的图像
process_images_in_zip(zip_path, output_folder)
```
